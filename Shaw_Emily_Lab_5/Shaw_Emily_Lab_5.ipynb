```python
from selenium import webdriver #imports webdriver and other packages that are used to allow the program to interact with the website
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import time
```

```python
PATH = Service("/Users/emilyshaw/Downloads/chromedriver_mac64/chromedriver") #sets the path to where the webdriver is located in my files
driver = webdriver.Chrome(service = PATH)

practice_site = "https://techwithtim.net" #sets the website that selenium will interact with
driver.get(practice_site) #opens a browser with the site
```

```python
link = driver.find_element(By.LINK_TEXT, "Python Programming") #locates the hyperlink that takes you to the page we want
link.click() #clicks the hyperlink

try: #using the clause avoids error messages unless there really is an error
    element = WebDriverWait(driver, 10).until(  #tells computer to wait before running so that an error doesn't occur 
    EC.presence_of_element_located(By.LINK_TEXT, "Beginner Python Tutorials")) #specifies which link will be interacted with
    element.click() #clicks the link specified before
    
    element = WebDriverWait(driver, 10).until(  #same as above
    EC.presence_of_element_located(By.ID, "sow-button-19310003")) #designates which button will be pressed
    element.click() #clicks button
    
    driver.back() #goes back a page
    driver.back()
    driver.back()
    
    driver.forward() #goes forward a page
    
except: #is an error occurs, driver will quit and close the browser
        driver.quit() 
```

```python
PATH = Service("/Users/emilyshaw/Downloads/chromedriver_mac64/chromedriver")
driver = webdriver.Chrome(service = PATH)

hoops = "https://hoopshype.com/" 
driver.get(hoops) 
```

```python
try:   
    a = ActionChains(driver) #found at https://www.tutorialspoint.com/how-can-i-perform-mouse-hover-action-in-selenium-python 
    menu = driver.find_element(By.ID, "menu-item-458969") #needed to hover over the menu for the players link to be available
    a.move_to_element(menu).perform()

    element2 = driver.find_element(By.ID, "menu-item-16")
    element2.click()
except:
    driver.close()
```

```python
csv_file = open("nba_salaries.csv", "w", newline = "", encoding = "utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["player", "salary"])

try:
    element = driver.find_element(By.CLASS_NAME, "pinned")
    rows = element.find_elements(By.TAG_NAME, "tr") 

    for i in rows [1:]:
        player = i.find_element(By.CLASS_NAME, "name")
        player= player.text
        print(player)

        salary = i.find_element(By.CLASS_NAME, "hh-salaries-sorted")
        salary = salary.get_attribute("data-value")
        print(salary)

        csv_writer.writerow([player, salary])
    csv_file.close()
except:
    print("wrong")
```
